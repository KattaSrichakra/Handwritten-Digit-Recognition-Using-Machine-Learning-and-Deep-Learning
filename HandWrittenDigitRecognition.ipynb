Section 7: Data Preprocessing
</span>
</div>

```python
# --- Feature Scaling (Normalization) ---

x_train_scaled = x_train / 255.0

x_test_scaled = x_test / 255.0

print("Training Data Shape :", x_train_scaled.shape)

print("\nTesting Data Shape :", x_test_scaled.shape)

print("\nMinimum Pixel Value :", x_train_scaled.min())

print("\nMaximum Pixel Value :", x_train_scaled.max())

print("\n✅ Feature Scaling Completed Successfully!")
```

```python
# --- Reshape Images for Traditional Machine Learning Models ---
# Since traditional machine learning models require two-dimensional input, each image must be flattened from 28 × 28 to 784 features.

x_train_ml = x_train_scaled.reshape(
    x_train_scaled.shape[0],
    28*28)

x_test_ml = x_test_scaled.reshape(
    x_test_scaled.shape[0],
    28*28)

print("Training Data Shape :", x_train_ml.shape)

print("\nTesting Data Shape :", x_test_ml.shape)

print("\n✅ Reshaping for Machine Learning Models Completed Successfully!")
```

```python
# --- One-Hot Encoding of Target Variable ---

y_train_dl = to_categorical(
    y_train,
    num_classes=10)

y_test_dl = to_categorical(
    y_test,
    num_classes=10)

print("Training Labels Shape :", y_train_dl.shape)

print("\nTesting Labels Shape :", y_test_dl.shape)

print("\n✅ One-Hot Encoding Completed Successfully!")
```

```python
# --- Reshape Images for CNN ---

x_train_cnn = x_train_scaled.reshape(x_train_scaled.shape[0],28,28,1)

x_test_cnn = x_test_scaled.reshape(x_test_scaled.shape[0],28,28,1)

print("Training Data Shape :", x_train_cnn.shape)

print("\nTesting Data Shape :", x_test_cnn.shape)

print("\n✅ CNN Data Preparation Completed Successfully!")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

- Data preprocessing is a crucial step in developing accurate machine learning and deep learning models.

- Pixel values were normalized to the range of 0 to 1 to improve training stability and accelerate model convergence.

- For traditional machine learning algorithms, each image was flattened into a one-dimensional vector containing 784 features.

- One-hot encoding was applied to the target labels to facilitate multiclass classification in Artificial Neural Networks and Convolutional Neural Networks.

- Finally, image data was reshaped into four-dimensional tensors to preserve spatial information required by Convolutional Neural Networks.

- These preprocessing steps ensure that the data is properly formatted and optimized for both traditional machine learning and deep learning approaches.

</p>

<div style="font-family:Georgia,
monospace;font-size:18px;font-weight:bold;">
<span style="color:#800000;">
Section 8: Traditional Machine Learning Model Development
</span>
</div>

```python
# --- Traditional Machine Learning Models Initialization ---

models = {

    "Logistic Regression":
        LogisticRegression(),

    "Decision Tree":
        DecisionTreeClassifier(),

    "Random Forest":
        RandomForestClassifier(),

    "K-Nearest Neighbors":
        KNeighborsClassifier(),

    "Support Vector Machine":
        SVC()
}

print("✅ All Models Initialized Successfully!")
```

```python
# --- Model Training and Evaluation ---

results = {}

for model_name, model in models.items():

    # Train Model

    model.fit(x_train_ml, y_train)

    # Predictions

    y_pred = model.predict(x_test_ml)

    # Accuracy

    acc = accuracy_score(y_test,y_pred)

    # Precision

    prec = precision_score(y_test,y_pred,average='weighted')

    # Recall

    rec = recall_score(y_test,y_pred,average='weighted')

    # F1 Score

    f1 = f1_score(y_test,y_pred,average='weighted')

    # Training Accuracy

    train_acc = accuracy_score(y_train,model.predict(x_train_ml))

    # Testing Accuracy

    test_acc = accuracy_score(y_test,y_pred)

    # Store Results

    results[model_name] = {

        "Accuracy": acc,

        "Precision": prec,

        "Recall": rec,

        "F1 Score": f1,

        "Train Accuracy": train_acc,

        "Test Accuracy": test_acc

    }

print("✅ Model Training and Evaluation Completed Successfully!")
```

```python
# --- Model Comparison Table ---

comparison = pd.DataFrame(results).T

comparison = comparison.sort_values(
    by='Accuracy',
    ascending=False)

comparison = comparison.round(4)

print("Model Comparison Table")
print("-"*30)

display(comparison)
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

Several traditional machine learning algorithms were trained and evaluated for handwritten digit recognition.

Among all traditional machine learning models, **Support Vector Machine (SVM)** achieved the highest classification accuracy of approximately **97.9%**, demonstrating excellent predictive capability and strong generalization performance.

K-Nearest Neighbors and Random Forest also produced high accuracy scores, while Logistic Regression and Decision Tree showed comparatively lower performance.

Decision Tree and Random Forest achieved perfect training accuracy, indicating the presence of overfitting.

Overall, Support Vector Machine emerged as the best-performing traditional machine learning model and provides a strong baseline for comparison with deep learning architectures such as Artificial Neural Networks (ANN) and Convolutional Neural Networks (CNN).

</p>

```python
# --- Accuracy Comparison Graph ---

plt.figure(figsize=(6,4))

sns.barplot(
    x=comparison.index,
    y=comparison['Accuracy'],
    palette='viridis')

plt.title("Traditional Machine Learning Models Accuracy Comparison")

plt.xticks(rotation=45)

plt.xlabel("")

plt.ylabel("Accuracy")

plt.grid(axis='y')

plt.show()
```

Section 9: Artificial Neural Network (ANN)
</span>

```python
# --- Build Artificial Neural Network (ANN) Model ---

ann_model = Sequential([

    Flatten(input_shape=(28, 28)),

    Dense(128,activation='relu'),

    Dense(64,activation='relu'),

    Dense(10,activation='softmax')])

print("✅ ANN Model Created Successfully!")
```

```python
# --- Compile ANN Model ---

ann_model.compile(

    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy'])

print("✅ ANN Model Compiled Successfully!")
```

```python
# --- Train ANN Model ---

history_ann = ann_model.fit(

    x_train_scaled,

    y_train_dl,

    epochs=10,

    batch_size=32,

    validation_split=0.2,

    verbose=1)

print("✅ ANN Model Training Completed Successfully!")
```

```python
# --- Evaluate ANN Model ---

ann_loss, ann_accuracy = ann_model.evaluate(

    x_test_scaled,

    y_test_dl,

    verbose=0)

print("ANN Testing Accuracy :", round(ann_accuracy,4))

print("\n✅ ANN Model Evaluation Completed Successfully!")
```

```python
# --- Predictions using ANN ---

y_pred_ann = ann_model.predict(x_test_scaled)

y_pred_ann = np.argmax(y_pred_ann,axis=1)
```

```python
# --- Classification Report ---

print("Classification Report")
print("-"*25)

print(classification_report(y_test, y_pred_ann))
```

```python
# --- Confusion Matrix Heatmap ---

cm_ann = confusion_matrix(y_test,y_pred_ann)

plt.figure(figsize=(6,4))

sns.heatmap(

    cm_ann,

    annot=True,

    fmt='d',

    cmap='Blues')

plt.xlabel("Predicted Label")

plt.ylabel("Actual Label")

plt.title("Confusion Matrix - ANN")

plt.show()

print("✅ Confusion Matrix Generated Successfully!")
```

```python
# --- Accuracy Curve ---

plt.figure(figsize=(6,4))

plt.plot(
    history_ann.history['accuracy'],
    label='Training Accuracy')

plt.plot(
    history_ann.history['val_accuracy'],
    label='Validation Accuracy')

plt.xlabel("Epochs")

plt.ylabel("Accuracy")

plt.title("ANN Accuracy Curve")

plt.legend()

plt.show()
```

```python
# --- Loss Curve ---

plt.figure(figsize=(6,4))

plt.plot(
    history_ann.history['loss'],
    label='Training Loss')

plt.plot(
    history_ann.history['val_loss'],
    label='Validation Loss')

plt.xlabel("Epochs")

plt.ylabel("Loss")

plt.title("ANN Loss Curve")

plt.legend()

plt.show()
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">
<b>Insight</b><br>

An Artificial Neural Network (ANN) consists of multiple fully connected layers and utilizes the ReLU activation function for hidden layers and the Softmax activation function for multiclass classification.

The Adam optimizer and categorical cross-entropy loss function were employed to train the network efficiently.

ANN models are capable of learning complex nonlinear relationships and generally provide better performance than traditional machine learning algorithms.

However, because ANN models operate on flattened image vectors, they do not explicitly preserve spatial information contained in images.

The Artificial Neural Network achieved a testing accuracy of approximately 97.3%, demonstrating excellent predictive performance for handwritten digit recognition.

The confusion matrix indicates that most digits were classified correctly, with only a few misclassifications among visually similar classes.

Training and validation accuracy curves show stable convergence and strong learning capability.

Although the training accuracy continued to improve throughout the epochs, the validation accuracy stabilized around 97%, indicating slight overfitting after the later epochs.

Overall, the Artificial Neural Network significantly outperformed several traditional machine learning models and provided robust multiclass classification performance.

However, because ANN models operate on flattened image vectors and do not explicitly preserve spatial information, Convolutional Neural Networks (CNNs) are expected to achieve even better performance for handwritten digit recognition tasks.

Section 10: Convolutional Neural Network (CNN)
</span>
</div>

```python
# --- Build CNN Model ---

cnn_model = Sequential([

    Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(28,28,1)),

    MaxPooling2D(pool_size=(2,2)),

    Conv2D(
        filters=64,
        kernel_size=(3,3),
        activation='relu'),

    MaxPooling2D(pool_size=(2,2)),

    Flatten(),

    Dense(128,activation='relu'),

    Dropout(0.3),

    Dense(10,activation='softmax')
])

print("✅ CNN Model Created Successfully!")
```

```python
# --- Compile CNN Model ---

cnn_model.compile(

    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy'])

print("✅ CNN Model Compiled Successfully!")
```

```python
# --- CNN Model Summary ---

cnn_model.summary()
```

```python
# --- Train CNN Model ---

history_cnn = cnn_model.fit(

    x_train_cnn,

    y_train_dl,

    epochs=10,

    batch_size=32,

    validation_split=0.2,

    verbose=1)

print("\n✅ CNN Model Training Completed Successfully!")
```

```python
# --- Evaluate CNN Model ---

cnn_loss, cnn_accuracy = cnn_model.evaluate(

    x_test_cnn,

    y_test_dl,

    verbose=0)

print("CNN Testing Accuracy :", round(cnn_accuracy,4))

print("\n✅ CNN Model Evaluation Completed Successfully!")
```

```python
# --- Predictions using CNN ---

y_pred_cnn = cnn_model.predict(x_test_cnn)

y_pred_cnn = np.argmax(y_pred_cnn,axis=1)

print("✅ Predictions Generated Successfully!")
```

```python
# --- Classification Report ---

print("Classification Report")
print("-"*25)

print(classification_report(y_test,y_pred_cnn))
```

```python
# --- Confusion Matrix Heatmap ---

cm_cnn = confusion_matrix(y_test,y_pred_cnn)

plt.figure(figsize=(6,4))

sns.heatmap(

    cm_cnn,

    annot=True,

    fmt='d',

    cmap='Blues')

plt.xlabel("Predicted Label")

plt.ylabel("Actual Label")

plt.title("Confusion Matrix - CNN")

plt.show()

print("\n✅ Confusion Matrix Generated Successfully!")
```

```python
# --- Training and Validation Accuracy Curve ---

plt.figure(figsize=(6,4))

plt.plot(

    history_cnn.history['accuracy'],

    label='Training Accuracy')

plt.plot(

    history_cnn.history['val_accuracy'],

    label='Validation Accuracy')

plt.xlabel("Epochs")

plt.ylabel("Accuracy")

plt.title("CNN Accuracy Curve")

plt.legend()

plt.grid()

plt.show()
```

```python
# --- Training and Validation Loss Curve ---

plt.figure(figsize=(6,4))

plt.plot(

    history_cnn.history['loss'],

    label='Training Loss')

plt.plot(

    history_cnn.history['val_loss'],

    label='Validation Loss')

plt.xlabel("Epochs")

plt.ylabel("Loss")

plt.title("CNN Loss Curve")

plt.legend()

plt.grid()

plt.show()
```

```python
# --- Visualize sample Predictions ---

plt.figure(figsize=(12,8))

for i in range(15):

    plt.subplot(3,5,i+1)

    plt.imshow(
        x_test[i],
        cmap='gray')

    plt.title(
        f"Predicted : {y_pred_cnn[i]}")

    plt.axis('off')

plt.tight_layout()

plt.show()

print("\n✅ Sample Predictions Visualized Successfully!")
```
